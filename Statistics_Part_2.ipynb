**Statistics Part 2**

1. What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method to make inferences about population parameters based on sample data, deciding whether to reject a null hypothesis using tests like Z-test or T-test. In Python, libraries like scipy.stats are used (e.g., scipy.stats.ttest_1samp).

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H₀) states no effect or difference (e.g., mean = 0), while the alternative hypothesis (H₁) states there is an effect or difference (e.g., mean ≠ 0). In Python, you define these when using tests like scipy.stats.ttest_ind.

3. What is the significance level in hypothesis testing, and why is it important?
  - The significance level (α) is the probability of rejecting H₀ when it’s true, typically 0.05 (5%). It’s important to control Type 1 errors. In Python, compare the p-value from scipy.stats tests to α.

4. What does a P-value represent in hypothesis testing?
  - The p-value measures the probability of observing the test results under H₀. A small p-value (e.g., < 0.05) suggests evidence against H₀. Python’s scipy.stats functions (e.g., ttest_ind) return p-values.

5. How do you interpret the P-value in hypothesis testing?
  - If p-value ≤ α (e.g., 0.05), reject H₀; if p-value > α, fail to reject H₀. In Python, after running scipy.stats.ttest_ind, compare the p-value to α to decide.

6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Type 1 error (α) is rejecting H₀ when it’s true (false positive). Type 2 error (β) is failing to reject H₀ when H₁ is true (false negative). Python doesn’t directly compute these, but they relate to α and power.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - A one-tailed test checks for an effect in one direction (e.g., mean > 0), while a two-tailed test checks both directions (e.g., mean ≠ 0). In Python, use alternative='greater' or alternative='two-sided' in scipy.stats.ttest_ind.

8. What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test compares a sample mean to a known population mean when the population variance is known and the sample size is large (n > 30). In Python, use scipy.stats.norm to compute Z-scores and p-values.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - Z-score = (sample mean - population mean) / (population std / √n). It measures how many standard deviations the sample mean is from the population mean. In Python, calculate using z = (x - mu) / (sigma / np.sqrt(n)).

10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The T-distribution is used when the sample size is small (n < 30) or population variance is unknown. It has heavier tails than the normal distribution. In Python, use scipy.stats.t for T-distribution calculations.

11. What is the difference between a Z-test and a T-test?
  - Z-test is used for large samples with known variance, while T-test is for small samples or unknown variance. In Python, use scipy.stats.norm for Z-tests and scipy.stats.ttest_1samp for T-tests.

12. What is the T-test, and how is it used in hypothesis testing?
  - The T-test compares sample means to test H₀ when variance is unknown or sample size is small. In Python, use scipy.stats.ttest_ind for two independent samples or ttest_1samp for one sample.

13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Z-test and T-test both test means, but Z-test uses normal distribution (large samples, known variance), while T-test uses T-distribution (small samples, unknown variance). Python implements both via scipy.stats.

14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval (CI) is a range estimating a population parameter with a confidence level (e.g., 95%). It’s used to assess precision of estimates. In Python, use scipy.stats.t.interval to compute a CI.

15. What is the margin of error, and how does it affect the confidence interval?
  - The margin of error (MoE) is the maximum expected difference between the true parameter and the estimate, affecting CI width. Larger MoE widens CI. In Python, MoE is part of scipy.stats.t.interval output.

16. How is Bayes’ Theorem used in statistics, and what is its significance?
  - Bayes’ Theorem updates probabilities based on new evidence: P(A|B) = P(B|A) * P(A) / P(B). It’s significant in Bayesian inference. In Python, use bayes_mvs from statsmodels.stats.descriptivestats for Bayesian stats.

17. What is the Chi-square distribution, and when is it used?
  - The Chi-square distribution is used for tests of independence or goodness of fit, often with categorical data. In Python, use scipy.stats.chi2 for Chi-square calculations.

18. What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness of fit test checks if observed frequencies match expected ones. Apply it to categorical data. In Python, use scipy.stats.chisquare to compare observed and expected frequencies.

19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution compares variances between two or more groups, used in ANOVA or F-tests. In Python, use scipy.stats.f for F-distribution calculations.

20. What is an ANOVA test, and what are its assumptions?
  - ANOVA tests if means of three or more groups differ. Assumptions: normality, equal variances, independent samples. In Python, use scipy.stats.f_oneway for ANOVA.

21. What are the different types of ANOVA tests?
  - Types: One-way ANOVA (one factor), Two-way ANOVA (two factors), and Repeated Measures ANOVA (same subjects). In Python, scipy.stats.f_oneway handles one-way ANOVA; use statsmodels for others.

22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test compares variances of two groups to test H₀ (equal variances). It’s used in ANOVA. In Python, compute with scipy.stats.f or as part of f_oneway.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm

# Sample data
sample = [23, 25, 22, 27, 21, 24, 26]
pop_mean = 20
pop_std = 3
alpha = 0.05

# Calculate Z-score
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed

# Output
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print("Reject H0" if p_value < alpha else "Fail to reject H0")
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
sample = [23, 25, 22, 27, 21, 24, 26]
pop_mean = 20
pop_std = 3
alpha = 0.05

# Z-test
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Critical values
z_critical = norm.ppf(1 - alpha/2)

# Plot
x = np.linspace(-4, 4, 100)
plt.plot(x, norm.pdf(x), label='Standard Normal')
plt.fill_between(x, norm.pdf(x), where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label='Rejection Region')
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.legend()
plt.title('Two-tailed Z-test Decision Region')
plt.show()

# Output
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print("Reject H0" if p_value < alpha else "Fail to reject H0")
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def type_errors(mu0, mu1, sigma, n, alpha):
    # Critical value for Type 1 error
    z_critical = norm.ppf(1 - alpha/2)
    se = sigma / np.sqrt(n)
    x_critical = mu0 + z_critical * se

    # Type 1 error (alpha)
    type1 = alpha

    # Type 2 error (beta)
    z_beta = (x_critical - mu1) / se
    type2 = norm.cdf(z_beta)

    # Plot
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 100)
    plt.plot(x, norm.pdf(x, mu0, se), label='H0: mu = 10')
    plt.plot(x, norm.pdf(x, mu1, se), label='H1: mu = 11')
    plt.fill_between(x, norm.pdf(x, mu0, se), where=(x >= x_critical), color='red', alpha=0.5, label='Type 1 Error')
    plt.fill_between(x, norm.pdf(x, mu1, se), where=(x <= x_critical), color='blue', alpha=0.5, label='Type 2 Error')
    plt.legend()
    plt.title('Type 1 and Type 2 Errors')
    plt.show()

    return type1, type2

# Parameters
mu0, mu1, sigma, n, alpha = 10, 11, 2, 30, 0.05
type1, type2 = type_errors(mu0, mu1, sigma, n, alpha)

# Output
print(f"Type 1 Error (alpha): {type1:.4f}")
print(f"Type 2 Error (beta): {type2:.4f}")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

# Function to calculate confidence interval
def confidence_interval(sample, confidence=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_error = t_critical * (sample_std / np.sqrt(n))
    return sample_mean - margin_error, sample_mean + margin_error

# Sample data
sample = [23, 25, 22, 27, 21, 24, 26]

# Calculate CI
ci_lower, ci_upper = confidence_interval(sample)

# Output
print(f"Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
```

```python
#11.Implement a Bayesian inference method using Bayes’ Theorem in Python and explain the process.
# Simple Bayesian inference for probability of event A given B
def bayes_theorem(p_a, p_b_given_a, p_b):
    return (p_b_given_a * p_a) / p_b

# Probabilities
p_a = 0.01  # Prior probability of A
p_b_given_a = 0.9  # Likelihood of B given A
p_b = 0.05  # Total probability of B

# Calculate posterior probability
posterior = bayes_theorem(p_a, p_b_given_a, p_b)

# Output
print(f"Posterior Probability: {posterior:.4f}")
```

```python
#16.Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
from scipy.stats import f

# Sample data
sample1 = [23, 25, 22, 27, 21]
sample2 = [30, 28, 32, 29, 31]

# F-test
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)
f_stat = var1 / var2
df1, df2 = len(sample1) - 1, len(sample2) - 1
p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

# Output
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print("Reject H0" if p_value < 0.05 else "Fail to reject H0")
```

```python
#19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    # Normality (Shapiro-Wilk test)
    normality = all(shapiro(group)[1] > 0.05 for group in groups)

    # Equal variance (Levene's test)
    equal_var = levene(*groups)[1] > 0.05

    return normality, equal_var

# Sample data
group1 = [23, 25, 22, 27, 21]
group2 = [30, 28, 32, 29, 31]
group3 = [26, 24, 27, 25, 23]

# Check assumptions
normality, equal_var = check_anova_assumptions(group1, group2, group3)

# Output
print(f"Normality Satisfied: {normality}")
print(f"Equal Variance Satisfied: {equal_var}")
```

```python
#20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt

# Sample data
data = pd.DataFrame({
    'Value': [23, 25, 22, 30, 28, 32, 26, 24, 27],
    'Factor1': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Factor2': ['X', 'X', 'Y', 'X', 'X', 'Y', 'X', 'Y', 'Y']
})

# Two-way ANOVA
model = ols('Value ~ Factor1 + Factor2 + Factor1:Factor2', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Plot interaction
data.groupby(['Factor1', 'Factor2']).mean()['Value'].unstack().plot(kind='bar')
plt.title('Two-way ANOVA Interaction Plot')
plt.show()

# Output
print(anova_table)
```

```python
#24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np
from scipy.stats import chi2

# Sample data
data = [23, 25, 22, 27, 21, 24, 26]
pop_var = 4  # Hypothesized population variance

# Sample variance
sample_var = np.var(data, ddof=1)
n = len(data)
chi2_stat = (n - 1) * sample_var / pop_var
p_value = 2 * min(chi2.cdf(chi2_stat, n-1), 1 - chi2.cdf(chi2_stat, n-1))

# Output
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print("Reject H0" if p_value < 0.05 else "Fail to reject H0")
```

```python
#25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
from scipy.stats import norm

# Data
n1, x1 = 100, 30  # Group 1: sample size, successes
n2, x2 = 120, 45  # Group 2: sample size, successes

# Proportions
p1 = x1 / n1
p2 = x2 / n2
p_pooled = (x1 + x2) / (n1 + n2)
se = np.sqrt(p_pooled * (1 - p_pooled) * (1/n1 + 1/n2))
z_score = (p1 - p2) / se
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print("Reject H0" if p_value < 0.05 else "Fail to reject H0")
```
